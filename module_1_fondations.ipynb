## 1. Variables et types de données

Une **variable** est une boîte dans laquelle on stocke une valeur. En Python, on n'a pas besoin de déclarer le type à l'avance : Python le détecte automatiquement.

Il existe 4 types de base :

| Type | Mot-clé | Exemple | Usage data analyst |
|------|---------|---------|--------------------|
| Entier | `int` | `42` | Quantités, IDs clients |
| Décimal | `float` | `19.99` | Prix, chiffres d'affaires |
| Texte | `str` | `"Paris"` | Noms, régions, catégories |
| Booléen | `bool` | `True` | Filtres, conditions |

````python
# ── Exemple 1 : Créer des variables pour une vente ──────────────────────────

id_commande = 1042           # int   → un nombre entier, l'identifiant de la commande
montant_ht  = 250.00         # float → un nombre décimal, le montant hors taxe
client      = "Dupont SA"    # str   → une chaîne de caractères, le nom du client
livree      = True           # bool  → un booléen, la commande est-elle livrée ?

# La fonction type() permet de vérifier le type d'une variable
print(type(id_commande))     # → <class 'int'>
print(type(montant_ht))      # → <class 'float'>
print(type(client))          # → <class 'str'>
print(type(livree))          # → <class 'bool'>
````

> 💡 **Bonne pratique** : nommer ses variables de façon explicite. Préférer `montant_ht` à `m` ou `x`. En data analyse, on manipule beaucoup de colonnes — des noms clairs évitent les erreurs.

## 2. Opérations de base

Python se comporte comme une calculatrice. Les opérateurs essentiels :

| Opérateur | Symbole | Exemple | Résultat |
|-----------|---------|---------|----------|
| Addition | `+` | `10 + 3` | `13` |
| Soustraction | `-` | `10 - 3` | `7` |
| Multiplication | `*` | `10 * 3` | `30` |
| Division | `/` | `10 / 3` | `3.333...` |
| Division entière | `//` | `10 // 3` | `3` |
| Modulo (reste) | `%` | `10 % 3` | `1` |
| Puissance | `**` | `10 ** 2` | `100` |

````python
# ── Exemple 2 : Calculer des indicateurs de vente ───────────────────────────

montant_ht  = 250.00    # Prix hors taxe
taux_tva    = 0.20      # TVA à 20 %
quantite    = 4         # Nombre d'articles commandés
remise      = 0.10      # Remise de 10 %

# Calcul de la TVA
montant_tva = montant_ht * taux_tva          # 250 × 0.20 = 50.0

# Calcul du montant TTC
montant_ttc = montant_ht + montant_tva       # 250 + 50 = 300.0

# Calcul du CA total (quantité × prix TTC)
ca_total = quantite * montant_ttc            # 4 × 300 = 1200.0

# Application de la remise commerciale
ca_apres_remise = ca_total * (1 - remise)    # 1200 × 0.90 = 1080.0

print(montant_tva)       # 50.0
print(montant_ttc)       # 300.0
print(ca_total)          # 1200.0
print(ca_apres_remise)   # 1080.0
````

## 3. Affichage formaté avec les f-strings

En data analyse, on affiche souvent des résultats lisibles. La méthode la plus moderne et la plus pratique en Python est le **f-string** (format string) : on préfixe la chaîne par `f` et on insère des variables entre `{}`.

````python
# ── Exemple 3 : Afficher un résumé de commande formaté ──────────────────────

client          = "Dupont SA"
id_commande     = 1042
ca_apres_remise = 1080.0

# Syntaxe f-string : f"texte {variable} texte"
print(f"Client : {client}")
print(f"Commande n°{id_commande}")

# :.2f → affiche le float avec exactement 2 décimales
print(f"Montant à facturer : {ca_apres_remise:.2f} €")

# On peut aussi faire des calculs directement dans le f-string
taux_tva = 0.20
print(f"Dont TVA : {ca_apres_remise * taux_tva / (1 + taux_tva):.2f} €")
````

````python
# ── Exemple 4 : Autres formatages utiles ────────────────────────────────────

valeur = 1234567.89

# Séparateur de milliers
print(f"{valeur:,.2f}")       # → 1,234,567.89

# Pourcentage
taux = 0.1567
print(f"{taux:.1%}")          # → 15.7%

# Alignement dans un tableau texte
print(f"{'Produit':<15} {'CA':>10}")       # Aligner à gauche / droite
print(f"{'Laptop':<15} {1250.00:>10.2f}")
print(f"{'Souris':<15} {  29.99:>10.2f}")
````

## 4. Manipulation des chaînes de caractères (str)

Les données textuelles sont omniprésentes en data analyse : noms de clients, catégories de produits, régions... Python offre de nombreuses méthodes pour les manipuler.

````python
# ── Exemple 5 : Nettoyer et transformer des données textuelles ───────────────

# Données "brutes" telles qu'elles pourraient arriver d'un fichier CSV
client_brut = "  dupont sa  "   # espaces en trop, minuscules

# .strip()  → supprime les espaces au début et à la fin
client_propre = client_brut.strip()
print(client_propre)             # → "dupont sa"

# .upper()  → tout en majuscules
# .lower()  → tout en minuscules
# .title()  → Première lettre de chaque mot en majuscule
print(client_propre.upper())     # → "DUPONT SA"
print(client_propre.title())     # → "Dupont Sa"

# .replace() → remplacer une sous-chaîne
region = "Ile-de-France"
print(region.replace("-", " "))  # → "Ile de France"

# .split()  → découper une chaîne en liste selon un séparateur
tags = "ventes;marketing;b2b"
print(tags.split(";"))           # → ['ventes', 'marketing', 'b2b']

# len()     → longueur d'une chaîne
print(len("Dupont SA"))          # → 9

# "in"      → vérifier si une sous-chaîne est présente
print("SA" in "Dupont SA")       # → True
````

## 5. Conversion de types (casting)

Les données issues de fichiers CSV arrivent souvent sous forme de **texte**, même les nombres. Il faut les convertir avant de faire des calculs.

````python
# ── Exemple 6 : Conversion de types ─────────────────────────────────────────

# Simulation : données lues depuis un CSV (tout est str par défaut)
quantite_str  = "4"
prix_str      = "250.50"
livree_str    = "True"

# int()   → convertit en entier
quantite = int(quantite_str)

# float() → convertit en décimal
prix = float(prix_str)

# bool()  → attention : bool("False") retourne True ! (toute str non vide = True)
# Pour les booléens depuis texte, mieux vaut une comparaison explicite
livree = livree_str == "True"    # → True si le texte est exactement "True"

# Maintenant on peut faire des calculs
total = quantite * prix
print(f"Total : {total:.2f} €")  # → Total : 1002.00 €

# str() → convertit n'importe quoi en texte (utile pour concaténer)
reference = "CMD-" + str(1042)   # Sans str(), Python lèverait une erreur
print(reference)                 # → CMD-1042
````

### Exercice 2 — Calcul de marge

Un commercial a réalisé une vente avec les données suivantes :
- Prix de vente HT : `450.00 €`
- Prix d'achat HT : `310.00 €`
- Quantité vendue : `7`

**À faire :**
1. Calculer la marge unitaire (prix vente - prix achat)
2. Calculer le taux de marge : `(marge unitaire / prix achat) × 100`
3. Calculer le CA total HT
4. Calculer la marge totale
5. Afficher un résumé avec ce format :
```
CA total : 3150.00 €
Marge unitaire : 140.00 €
Taux de marge : 45.2%
Marge totale : 980.00 €
```

````python
# Exercice 2 — Votre code ici
# 1. Calculer la marge unitaire (prix vente - prix achat)
prix_vente_ht   = 450.00
prix_achat_ht   = 310.00
qtt_vente       = 7
marge_unit      = prix_vente_ht - prix_achat_ht

# 2. Calculer le taux de marge
taux_marge = marge_unit / prix_achat_ht 

# 3. Calculer le CA total HT
ca_ht_total = prix_vente_ht * qtt_vente

# 4. Calculer la marge totale
marge_totale = (prix_vente_ht * qtt_vente) - (prix_achat_ht * qtt_vente)

# 5. Afficher un résumé
print("5. Afficher un résumé")
print(f"CA total : {ca_ht_total:,.2f} €")
print(f"Marge unitaire : {marge_unit} €")
print(f"Taux de marge : {taux_marge:.1%}")
print(f"Marge totale : {marge_totale:.2f} €")
````
