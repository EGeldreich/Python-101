## 1. Les listes — stocker des séries de données

Une **liste** est une collection ordonnée de valeurs. C'est la structure la plus courante pour stocker une série de données : une colonne de chiffres, une liste de clients, une série de mois.

- Elle s'écrit entre crochets `[ ]`
- Elle peut contenir des types différents
- Elle est **ordonnée** : chaque élément a un index qui commence à `0`
- Elle est **modifiable** : on peut ajouter, supprimer, modifier des éléments

````python
# ── Exemple 1 : Créer et accéder à une liste ────────────────────────────────

# Une liste de chiffres d'affaires mensuels (janvier à juin)
ca_mensuel = [12500, 14200, 9800, 16100, 13750, 15400]

# Accès par index — l'index commence à 0
print(ca_mensuel[0])     # → 12500  (janvier)
print(ca_mensuel[2])     # → 9800   (mars)

# Index négatif : on part de la fin
print(ca_mensuel[-1])    # → 15400  (dernier mois, juin)
print(ca_mensuel[-2])    # → 13750  (avant-dernier, mai)

# Longueur de la liste
print(len(ca_mensuel))   # → 6

# Slicing : extraire une sous-liste [début:fin] (fin exclue)
print(ca_mensuel[0:3])   # → [12500, 14200, 9800]  (3 premiers mois)
print(ca_mensuel[3:])    # → [16100, 13750, 15400] (du 4ème jusqu'à la fin)
````

````python
# ── Exemple 2 : Modifier une liste ──────────────────────────────────────────

clients = ["Dupont SA", "Martin Logistics", "Leclerc & Co"]

# Ajouter un élément à la fin
clients.append("Bernard Industries")
print(clients)    # → ['Dupont SA', 'Martin Logistics', 'Leclerc & Co', 'Bernard Industries']

# Modifier un élément existant
clients[1] = "Martin Transport"
print(clients)    # → ['Dupont SA', 'Martin Transport', 'Leclerc & Co', 'Bernard Industries']

# Supprimer un élément par sa valeur
clients.remove("Leclerc & Co")
print(clients)    # → ['Dupont SA', 'Martin Transport', 'Bernard Industries']

# Insérer à une position précise : insert(index, valeur)
clients.insert(1, "Petit & Fils")
print(clients)    # → ['Dupont SA', 'Petit & Fils', 'Martin Transport', 'Bernard Industries']
````

````python
# ── Exemple 3 : Fonctions utiles sur les listes numériques ──────────────────

ca_mensuel = [12500, 14200, 9800, 16100, 13750, 15400]

print(sum(ca_mensuel))     # → 81750  — somme totale
print(min(ca_mensuel))     # → 9800   — valeur minimale
print(max(ca_mensuel))     # → 16100  — valeur maximale

# Moyenne : somme divisée par le nombre d'éléments
moyenne = sum(ca_mensuel) / len(ca_mensuel)
print(f"CA moyen mensuel : {moyenne:.2f} €")   # → 13625.00 €

# Trier une liste (retourne une nouvelle liste triée)
print(sorted(ca_mensuel))               # → ordre croissant
print(sorted(ca_mensuel, reverse=True)) # → ordre décroissant

# Vérifier si une valeur est dans la liste
print(9800 in ca_mensuel)    # → True
print(5000 in ca_mensuel)    # → False
````

## 2. Les dictionnaires — structurer des données clients

Un **dictionnaire** stocke des données sous forme de paires **clé : valeur**. C'est la structure idéale pour représenter un enregistrement structuré : une fiche client, une ligne de commande, un produit.

- Il s'écrit entre accolades `{ }`
- On accède aux valeurs par leur **clé**, pas par un index numérique
- Les clés sont généralement des chaînes de caractères
- Il est **modifiable** : on peut ajouter ou modifier des clés

````python
# ── Exemple 4 : Créer et accéder à un dictionnaire ──────────────────────────

# Fiche d'un client
client = {
    "id"       : 1042,
    "nom"      : "Dupont SA",
    "region"   : "Grand Est",
    "ca_annuel": 125000.0,
    "actif"    : True
}

# Accès à une valeur par sa clé
print(client["nom"])        # → Dupont SA
print(client["ca_annuel"])  # → 125000.0

# .get() : accès sécurisé — retourne None si la clé n'existe pas (évite une erreur)
print(client.get("email"))              # → None
print(client.get("email", "inconnu"))   # → "inconnu" (valeur par défaut)

# Ajouter ou modifier une clé
client["email"] = "contact@dupont.fr"   # ajout d'une nouvelle clé
client["ca_annuel"] = 132000.0          # modification d'une clé existante

print(client)
````

````python
# ── Exemple 5 : Parcourir un dictionnaire ───────────────────────────────────

client = {
    "nom"      : "Dupont SA",
    "region"   : "Grand Est",
    "ca_annuel": 125000.0
}

# .keys()   → toutes les clés
print(client.keys())     # → dict_keys(['nom', 'region', 'ca_annuel'])

# .values() → toutes les valeurs
print(client.values())   # → dict_values(['Dupont SA', 'Grand Est', 125000.0])

# .items()  → paires (clé, valeur) — très utile pour les boucles (voir section 5)
print(client.items())
````

````python
# ── Exemple 6 : Liste de dictionnaires — la structure la plus utilisée ───────
# C'est exactement ce que représente un tableau/fichier CSV en mémoire

ventes = [
    {"id": 1, "client": "Dupont SA",         "montant": 1200.0, "region": "Grand Est"},
    {"id": 2, "client": "Martin Logistics",   "montant": 3400.0, "region": "Île-de-France"},
    {"id": 3, "client": "Bernard Industries", "montant": 870.0,  "region": "Grand Est"},
    {"id": 4, "client": "Petit & Fils",       "montant": 2100.0, "region": "Occitanie"}
]

# Accéder à la 2ème vente (index 1)
print(ventes[1])

# Accéder au montant de la 2ème vente
print(ventes[1]["montant"])    # → 3400.0

# Nombre de ventes
print(len(ventes))             # → 4
````

## 3. Tuples et ensembles — usage rapide

Ces deux structures sont moins centrales en data analyse, mais il faut les connaître.

````python
# ── Exemple 7 : Tuple — liste immuable ──────────────────────────────────────
# Un tuple s'écrit avec des parenthèses. Il ne peut pas être modifié après création.
# Usage typique : stocker des constantes, des coordonnées, des paires fixes.

trimestres = ("T1", "T2", "T3", "T4")   # toujours 4 trimestres, ça ne change pas
print(trimestres[0])    # → T1
# trimestres[0] = "Q1"  # → ceci provoquerait une erreur : tuple non modifiable

# Unpacking : affecter plusieurs variables d'un coup
t1, t2, t3, t4 = trimestres
print(t3)    # → T3


# ── Exemple 8 : Ensemble (set) — valeurs uniques ─────────────────────────────
# Un set ne conserve que les valeurs uniques et n'a pas d'ordre.
# Usage typique : trouver des valeurs distinctes dans une colonne.

regions_ventes = ["Grand Est", "Île-de-France", "Grand Est", "Occitanie", "Île-de-France"]

regions_uniques = set(regions_ventes)   # supprime les doublons
print(regions_uniques)                  # → {'Grand Est', 'Île-de-France', 'Occitanie'}
print(len(regions_uniques))             # → 3 régions distinctes
````

## 4. Conditions — if / elif / else

Les **conditions** permettent d'exécuter du code différent selon la valeur d'une variable. En data analyse, on les utilise constamment pour classer, filtrer ou signaler des données.

**Opérateurs de comparaison :**

| Opérateur | Signification |
|-----------|---------------|
| `==` | égal à |
| `!=` | différent de |
| `>` | strictement supérieur |
| `>=` | supérieur ou égal |
| `<` | strictement inférieur |
| `<=` | inférieur ou égal |
| `and` | les deux conditions vraies |
| `or` | au moins une condition vraie |
| `not` | inverse la condition |

````python
# ── Exemple 9 : Classer un client selon son CA ──────────────────────────────

ca_client = 87500

# Structure if / elif / elif / else
# Python exécute le premier bloc dont la condition est True, puis ignore les suivants

if ca_client >= 100000:
    segment = "Grand compte"       # condition vraie → ce bloc s'exécute
elif ca_client >= 50000:
    segment = "Compte intermédiaire"
elif ca_client >= 10000:
    segment = "PME"
else:
    segment = "Petit compte"       # aucune condition précédente n'était vraie

print(f"CA : {ca_client:,} € → Segment : {segment}")
# → CA : 87,500 € → Segment : Compte intermédiaire
````

````python
# ── Exemple 10 : Conditions combinées avec and / or ─────────────────────────

client_actif = True
ca_client    = 87500
retard_paiement = False

# and : les DEUX conditions doivent être vraies
if client_actif and ca_client >= 50000:
    print("Éligible programme fidélité")    # → s'affiche

# or : AU MOINS UNE condition doit être vraie
if ca_client < 10000 or retard_paiement:
    print("Alerte : vérifier ce compte")
else:
    print("Compte en bonne santé")          # → s'affiche

# not : inverse la condition
if not retard_paiement:
    print("Aucun retard de paiement")       # → s'affiche
````

## 5. Boucles — for et while

Une **boucle** permet de répéter une action sur chaque élément d'une collection, ou jusqu'à ce qu'une condition soit fausse. C'est la base de tout traitement automatisé de données.

### La boucle `for` — parcourir une collection

C'est la boucle la plus utilisée en data analyse : on l'applique sur chaque ligne d'un dataset, chaque valeur d'une liste, etc.

````python
# ── Exemple 11 : Boucle for sur une liste ───────────────────────────────────

ca_mensuel = [12500, 14200, 9800, 16100, 13750, 15400]
mois       = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun"]

# Parcourir une liste simple
for ca in ca_mensuel:
    print(ca)    # affiche chaque valeur une par une

print("---")

# enumerate() : parcourir avec l'index en même temps
for i, ca in enumerate(ca_mensuel):
    print(f"{mois[i]} : {ca:,} €")

print("---")

# zip() : parcourir deux listes en parallèle
for mois_nom, ca in zip(mois, ca_mensuel):
    print(f"{mois_nom} : {ca:,} €")
````

````python
# ── Exemple 12 : Boucle for sur une liste de dictionnaires ──────────────────
# Pattern fondamental en data analyse

ventes = [
    {"id": 1, "client": "Dupont SA",         "montant": 1200.0, "region": "Grand Est"},
    {"id": 2, "client": "Martin Logistics",   "montant": 3400.0, "region": "Île-de-France"},
    {"id": 3, "client": "Bernard Industries", "montant": 870.0,  "region": "Grand Est"},
    {"id": 4, "client": "Petit & Fils",       "montant": 2100.0, "region": "Occitanie"}
]

# Calculer le CA total en parcourant toutes les ventes
ca_total = 0
for vente in ventes:               # à chaque itération, vente = un dictionnaire
    ca_total += vente["montant"]   # += est un raccourci pour ca_total = ca_total + ...

print(f"CA total : {ca_total:,.2f} €")    # → CA total : 7,570.00 €

# Filtrer : afficher uniquement les ventes du Grand Est
print("\nVentes Grand Est :")
for vente in ventes:
    if vente["region"] == "Grand Est":    # condition à l'intérieur de la boucle
        print(f"  {vente['client']} — {vente['montant']:,.2f} €")
````

````python
# ── Exemple 13 : Boucle for sur un dictionnaire avec .items() ───────────────

ca_par_region = {
    "Grand Est"     : 42000,
    "Île-de-France" : 89500,
    "Occitanie"     : 31200,
    "Bretagne"      : 27800
}

total = sum(ca_par_region.values())

print(f"{'Région':<20} {'CA':>12} {'Part':>8}")
print("-" * 42)

# .items() retourne des paires (clé, valeur) à chaque itération
for region, ca in ca_par_region.items():
    part = ca / total             # proportion du CA total
    print(f"{region:<20} {ca:>12,} {part:>8.1%}")
````

````python
# ── Exemple 14 : range() — boucle sur une séquence de nombres ───────────────

# range(n)        → de 0 à n-1
# range(a, b)     → de a à b-1
# range(a, b, pas) → de a à b-1 par pas

# Simuler 5 mois de CA avec une évolution de +3% par mois
ca_base = 10000
for mois in range(1, 6):             # 1, 2, 3, 4, 5
    ca = ca_base * (1.03 ** mois)    # ** = puissance
    print(f"Mois {mois} : {ca:,.2f} €")
````

````python
# ── Exemple 15 : Boucle while — répéter jusqu'à une condition ───────────────
# Moins fréquente que for, mais utile pour des processus itératifs.

# Calculer combien de mois pour dépasser un objectif de CA cumulé
ca_mensuel_moyen = 13500
objectif         = 100000
ca_cumule        = 0
mois             = 0

while ca_cumule < objectif:        # tant que l'objectif n'est pas atteint
    mois     += 1
    ca_cumule += ca_mensuel_moyen

print(f"Objectif de {objectif:,} € atteint en {mois} mois")
print(f"CA cumulé  : {ca_cumule:,} €")
````

````python
# ── Exemple 16 : break et continue — contrôler une boucle ───────────────────

commandes = [
    {"id": 101, "montant": 500,  "statut": "livrée"},
    {"id": 102, "montant": 1200, "statut": "livrée"},
    {"id": 103, "montant": 340,  "statut": "annulée"},
    {"id": 104, "montant": 780,  "statut": "livrée"},
    {"id": 105, "montant": 920,  "statut": "livrée"}
]

# continue : passer à l'itération suivante sans exécuter la suite du bloc
print("Commandes livrées uniquement :")
ca_livre = 0
for cmd in commandes:
    if cmd["statut"] == "annulée":
        continue                    # on saute cette commande et on passe à la suivante
    ca_livre += cmd["montant"]
    print(f"  CMD-{cmd['id']} : {cmd['montant']} €")

print(f"  Total livré : {ca_livre} €")

print("\n---\n")

# break : stopper la boucle dès qu'une condition est rencontrée
print("Recherche première commande > 1000 € :")
for cmd in commandes:
    if cmd["montant"] > 1000:
        print(f"  Trouvée → CMD-{cmd['id']} : {cmd['montant']} €")
        break                       # on arrête dès qu'on a trouvé
````

## 6. List Abrégée — écrire une boucle en une ligne

C'est une syntaxe Python très utilisée par les data analysts pour transformer ou filtrer une liste de façon compacte.

````python
# ── Exemple 17 : List abrégée ─────────────────────────────────────────

montants_ht = [1000, 2500, 875, 3200, 640]

# Version boucle classique : calculer le TTC pour chaque montant
montants_ttc_v1 = []
for m in montants_ht:
    montants_ttc_v1.append(m * 1.20)

# Version list abrégée : [expression for element in liste]
montants_ttc_v2 = [m * 1.20 for m in montants_ht]

print(montants_ttc_v1)    # → [1200.0, 3000.0, 1050.0, 3840.0, 768.0]
print(montants_ttc_v2)    # → identique

# Avec une condition : [expression for element in liste if condition]
# Extraire uniquement les montants HT > 1000
grands_comptes = [m for m in montants_ht if m > 1000]
print(grands_comptes)     # → [2500, 3200]

# Extraire les noms de clients d'une région spécifique
ventes = [
    {"client": "Dupont SA",         "region": "Grand Est"},
    {"client": "Martin Logistics",  "region": "Île-de-France"},
    {"client": "Bernard Industries","region": "Grand Est"}
]
clients_grand_est = [v["client"] for v in ventes if v["region"] == "Grand Est"]
print(clients_grand_est)  # → ['Dupont SA', 'Bernard Industries']
````

````python
# ── CORRECTION Exercice 4 ────────────────────────────────────────────────────

# 1. Construire le dictionnaire CA par région
ca_par_region = {}

for client in clients:
    region = client["region"]
    ca     = client["ca"]

    if region in ca_par_region:          # si la région existe déjà dans le dict
        ca_par_region[region] += ca      # on additionne
    else:
        ca_par_region[region] = ca       # sinon on crée l'entrée

# 2. Calculer le total et afficher le classement
total = sum(ca_par_region.values())

# sorted() sur .items() avec une fonction lambda comme clé de tri
# lambda x: x[1] signifie "trier par le 2ème élément de la paire (clé, valeur)"
regions_triees = sorted(ca_par_region.items(), key=lambda x: x[1], reverse=True)

print("=== CA par région (classement décroissant) ===")
for region, ca in regions_triees:
    part = ca / total
    print(f"{region:<20} → {ca:>9,} €  ({part:.1%})")
````
